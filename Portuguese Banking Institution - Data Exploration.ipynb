```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bank_p = 'clean_bank_data.csv'
```

```python
bankcopy = pd.read_csv(bank_p, sep = ',', decimal = '.', header=0)
```

```python
bankcopy
```

```python
bankcopy['Age'].plot(kind='hist',bins=20, figsize=(6,6))
plt.title('Bank data - Age info')
plt.xlabel('Age')
plt.show()
```

```python
bankcopy_1 = bankcopy.copy()
p_fake = bankcopy_1[bankcopy_1['pdays'] == 999]
bankcopy_1 = bankcopy_1.drop(p_fake.index, axis = 0)
```

```python
bankcopy_1['pdays'].plot(kind='hist',bins=20, figsize=(8,8))
plt.title('Bank data - No. of days passed since last contacted in previous campaign')
plt.xlabel('Days passed since last campaign')
plt.show()
```
